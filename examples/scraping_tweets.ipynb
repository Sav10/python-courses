#### Recherche les Tweets du compte RERB

```python
# Creating list to append tweet data to
tweets_list1 = []
```

```python
# Using TwitterSearchScraper to scrape data and append tweets to list
for i,tweet in enumerate(sntwitter.TwitterSearchScraper('from:RERB').get_items()):
    if i>200:
        break
    tweets_list1.append([tweet.date, tweet.id, tweet.content, tweet.user.username])
```

```python
# Creating a dataframe from the tweets list above 
tweets_df1 = pd.DataFrame(tweets_list1, columns=['Datetime', 'Tweet Id', 'Text', 'Username'])
```

#### Recherche des 50 derniers tweets en France contenant le mot Covid

```python
keyword = '(COVID)'
maxTweets = 50
tweets = []
tdf = None
for i,tweet in enumerate(sntwitter.TwitterSearchScraper(keyword + ' since:2022-03-01 lang:fr').get_items()) :
        if i > maxTweets :
            break
        username = tweet.username
        text = tweet.content
        pubdate = tweet.date
        permalink = tweet.url
        tweets.append({
            "permalink":permalink,
            "pubdate":pubdate,
            "text":text,
            "username":username
        })
```
